1. What is the difference between a function and a method in Python?
   
* Definition:

 - Function: A standalone block of code that performs a specific task, defined using the def keyword.
 -Method: A function that is associated with an object and defined within a class.
* Association:

 - Function: Exists independently of classes and can be defined at the module level.
 - Method: Tied to class instances and operates within the context of a class.
* Invocation:

 - Function: Called directly by its name, e.g., square(4).
 - Method: Called on an instance of a class, e.g., rect.area().
* Access to Data:

 - Function: Does not have access to instance data unless explicitly passed as arguments.
 - Method: Can access and modify instance attributes (data) of the object it belongs to.
* Use Cases:

 - Function: Used for general-purpose tasks that do not require object context.
 - Method: Defines behaviors specific to the objects created from a class, encapsulating functionality related to that object.

5.  What are iterators in Python and how do they differ from iterables?

* Iterators
 - An iterator is an object in Python that implements the iterator protocol, which consists of the methods __iter__() and __next__().
 - An iterator allows you to traverse through a collection of elements (like lists, tuples, or dictionaries) one at a time without needing to know the underlying structure of the collection.
 - When you call __next__(), it returns the next item in the sequence. If there are no more items to return, it raises a StopIteration exception.

* Iterables
 - An iterable is any Python object that can return an iterator using the __iter__() method. This includes data types like lists, tuples, strings, and dictionaries.
 - An iterable can be looped over using a for loop or can be converted into an iterator using the iter() function.

* Key Differences

 -Definition:

  - An iterable is an object that can be iterated over (e.g., lists, strings).
  - An iterator is an object that keeps track of the current position during iteration and provides the next value.
* Methods:

 - Iterables implement the __iter__() method, which returns an iterator.
 - Iterators implement both __iter__() (which returns the iterator itself) and __next__() (which returns the next item).
* State:

 - An iterable can be iterated over multiple times, creating a new iterator each time.
 - An iterator maintains its state and can only be traversed once. After it has been exhausted, it cannot be reused.

6.  Explain the concept of generators in Python and how they are defined.

* Concept of Generators in Python

 - Generators are a special type of iterable in Python that allow for the creation of iterators in a more concise and memory-efficient way. They are defined using a function that contains one or more yield statements instead of return statements.
* Key Features of Generators:
 - Lazy Evaluation: Generators compute values on-the-fly and yield them one at a time, which means they do not store the entire output in memory. This is particularly useful for working with large datasets or streams of data.

 - State Retention: When a generator function is called, it does not execute the function body immediately. Instead, it returns a generator object that can be iterated over. Each time the next() function is called on the generator, execution resumes from where it last yielded a value, allowing it to maintain its state between calls.

 - Simplified Code: Generators can simplify code that would otherwise require complex iterator classes. They provide a clean and readable way to create iterators.

* How Generators are Defined:
 - Generators are defined using the def keyword, just like regular functions, but they use the yield statement to produce values.

* Using the Generator:
 - To use a generator, you can create a generator object by calling the generator function and then iterate over it using a for loop or the next() function.

8. What is a lambda function in Python and when is it typically used?

* Lambda Function in Python

 - A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with the def keyword, lambda functions are typically used for short, simple operations and can take any number of arguments but can only have a single expression.

* Key Characteristics of Lambda Functions:
 - Anonymous: Lambda functions do not have a name unless assigned to a variable.
 - Single Expression: They can only contain a single expression, which is evaluated and returned. They cannot contain statements or multiple expressions.

* Typical Use Cases for Lambda Functions:
 - Short Functions: Used for simple operations that are not reused elsewhere, making the code more concise.
 - Higher-Order Functions: Often used as arguments to higher-order functions like map(), filter(), and reduce().
 - Sorting: Used as a key function in sorting operations, especially with lists of tuples or dictionaries.
 - Event Handling: Commonly used in GUI programming or callback functions where a simple function is needed for a short duration.

* Key Features of map()
 - Function Application: map() takes two primary arguments: a function and an iterable. The function is applied to each element of the iterable.

 - Multiple Iterables: map() can accept multiple iterables as arguments. In this case, the function must take as many arguments as there are iterables.

* Usage of the map() Function in Python

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
   - The map(), reduce(), and filter() functions are all higher-order functions in Python that operate on iterables, but they serve different purposes and have distinct behaviors. Here’s a breakdown of their differences:

* map() Function
 - Purpose: Applies a specified function to each item in an iterable and returns an iterator of the results.
 - Input: A function and one or more iterables.
 - Output: An iterator containing the results of applying the function to each element.
 - Use Case: Transforming data (e.g., squaring numbers, converting data types).   

* reduce() Function
 - Purpose: Applies a specified function cumulatively to the items of an iterable, reducing the iterable to a single value.
 - Input: A function and an iterable.
 - Output: A single value that results from the cumulative application of the function.
 -Use Case: Aggregating data (e.g., summing numbers, finding the product).

* filter() Function
 - Purpose: Filters elements from an iterable based on a specified function that returns a boolean value.
 - Input: A function and an iterable.
 - Output: An iterator containing only the elements for which the function returns True.
 - Use Case: Selecting specific items from a collection (e.g., filtering even numbers).

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
my_list = [10, 15, 20, 25, 30]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)
```

```python
# 4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")

```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
# 6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for value in powers_of_two(5):
    print(value)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 'banana'), (2, 'apple'), (3, 'cherry'), (4, 'date')]

sorted_data = sorted(data, key=lambda x: x[1])

print("Sorted list of tuples:", sorted_data)
```

```python
# 9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)

```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))

# Example usage
input_str = "Hello, World!"
result = remove_vowels(input_str)
print("String without vowels:", result)
```

```python
# 11.  Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
"""Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map."""

orders = [
    [1, 20, 3],  # Total = 60
    [2, 15, 5],  # Total = 75
    [3, 50, 2],  # Total = 100
    [4, 5, 10],   # Total = 50
]

def calculate_total(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity

    if total_price < 100:
        total_price += 10
    return (order_number, total_price)

result = list(map(calculate_total, orders))

print("Order totals:")
for order in result:
    print(order)
```
